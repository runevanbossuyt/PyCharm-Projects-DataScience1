| f | 2 | 3 | 5 | 6 | 4 | 2 | 3

1. Welke percentielscore hoort bij score 8?

## Vraag 4

Een inkomensverdeling is als volgt

| categorie | A | B | C | D |
| --- | --- | --- | --- | --- |
| inkomen | 1000 tot 2000 | 2000 tot 3000 | 3000 tot 4000 | 4000 tot 5000 |
| frequentie | 30 | 40 | 30 | 45 |

1. Vervolledig de frequentietabel met relatieve frequenties (in percent), cumulatief percentage en percentielscores.

```python
# Relatieve Frequenties in %
# Om de relatieve frequenties te berekenen, delen we elke frequentie door de totale frequentie en vermenigvuldigen
# we het resultaat met 100 om het als een percentage te krijgen.
frequentie = pd.array([30, 40, 30, 45])

RFA = 30/frequentie.sum()
RFB = 40/frequentie.sum()
RFC = 30/frequentie.sum()
RFD = 45/frequentie.sum()

# in %
RFA *= 100
RFB *= 100
RFC *= 100
RFD *= 100

# Cumulatief Percentage
# Om de cumulatieve frequentie te berekenen, tellen we de frequenties op van alle categorieën
# die vóór de huidige categorie komen, inclusief de frequentie van de huidige categorie.
CFA = 30
CFB = 30 + 40
CFC = 30 + 40 + 30
CFD = 30 + 40 + 30 + 45

# Om de cumulatieve percentages te berekenen, delen we de cumulatieve frequentie
# van elke categorie door de totale frequentie en vermenigvuldigen we het resultaat met 100.
CPA = CFA/frequentie.sum() * 100
CPB = CFB/frequentie.sum() * 100
CPC = CFC/frequentie.sum() * 100
CPD = CFD/frequentie.sum() * 100

# Percentielscores
# Om de percentielscores te berekenen, nemen we het cumulatieve percentage van elke categorie en ronden we het af naar de dichtstbijzijnde gehele waarde.
PA = int(CPA.round())
PB = int(CPB.round())
PC = int(CPC.round())
PD = int(CPD.round())


# New Data
frequentietabel = pd.DataFrame({
    'A': ['1000 to 2000', 30, RFA, CPA, PA],
    'B': ['2000 to 3000', 40, RFB, CPB, PB],
    'C': ['3000 to 4000', 30, RFC, CPC, PC],
    'D': ['4000 to 5000', 45, RFD, CPD, PD]
})
display(frequentietabel)
```

## Vraag 9

gegeven volgende frequentietabel

| score | 4 | 5 | 6 | 7 | 8 | 9 | 10 | 11
| --- | --- | --- | --- |--- |---  |--- |--- |--- |
| frequentie | 5 | 7 | 9 | 11 | 12 | 10 | 8 | 6

1. Wat is de percentielscore die hoort bij score 9?
1. Wat is de percentielscore die hoort bij score 7.5?
1. Wat is de score die hoort bij percentielscore 31?

```python
score = [4, 5, 6, 7, 8, 9, 10, 11]
frequentie = [5,7, 9, 11, 12, 10, 8, 6]

df = pd.DataFrame({'score':score, 'frequentie':frequentie})

# Vraag 1 (Wat is de percentielscore die hoort bij score 9?)
cum_frequentie = df[df['score'] <= 9]['frequentie'].sum()
totaal_frequentie = df['frequentie'].sum()
cum_percentage = cum_frequentie / totaal_frequentie * 100

print(f"Cumulatief percentage bij score 9: {cum_percentage:.2f}%")

# Vraag 2 (Wat is de percentielscore die hoort bij score 7.5?)
cum_frequentie = df[df['score'] <= 7.5]['frequentie'].sum()
totaal_frequentie = df['frequentie'].sum()
cum_percentage = cum_frequentie / totaal_frequentie * 100

print(f"Cumulatief percentage bij score 7.5: {cum_percentage:.2f}%")

# Vraag 3 (Wat is de score die hoort bij percentielscore 31?)
target_percentage = 31
df['cum_percentage'] = df['frequentie'].cumsum() / df['frequentie'].sum() * 100
closest_score = df.loc[(df['cum_percentage'] - target_percentage).abs().idxmin(), 'score']

print(f"Score bij en cumulatief percentage van 31: {closest_score:d}")
```
