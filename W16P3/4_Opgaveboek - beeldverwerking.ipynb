### Maak een plot van de roodwaarden van het beeld
- Maak een kopie van het ingeladen bestand
- Zet de waarden van R en G op 0
- Plot vervolgens het verkregen beeld met de functie `imshow` van de matplotlib library.

```python
# (1)
kopieImg = arrayImg
```

```python
# (2)
kopieImg[:,:,0] = 0
kopieImg[:,:,1] = 0
```

```python
# (3)
# Toon de bewerkte afbeelding
Image.fromarray(kopieImg).show()
```

### Histogram van de helderheden
- Bereken nu de helderheid voor elke pixel m.b.v de onderstaande formule
- $ helderheid = 0.3 \cdot rood+0.59 \cdot groen+0.11 \cdot blauw $
- maak een histogram van de helderheden

```python
# helderheid = []
# for r in arrayImg[:,:,0]:
#     for g in arrayImg[:,:,1]:
#         for b in arrayImg[:,:,2]:
#             helderheid.append(0.3 * r + 0.59 * g + 0.11 * b)

helderheid = 0.3 * arrayImg[:,:,0] + 0.59 * arrayImg[:,:,1] + 0.11 * arrayImg[:,:,2]

print(helderheid) # TEST
```

```python
hist, bins = np.histogram(helderheid, bins=256, range=(0,255))

plt.bar(bins[:-1], hist, width=1)
plt.show()
```

### Maak een frequentiepolygoon van de helderheden

```python
fig,axes = plt.subplots()

_=axes.hist(helderheid,edgecolor='black',label='Helderheid')
_=axes.set_title('Helderheid')
_=axes.set_xlabel('Kwantiteit')
_=axes.set_ylabel('Helderheid')
_=axes.legend()
```
